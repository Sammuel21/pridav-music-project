# Data Sample Structure

This notebook shows sample data structure from our Spotify dataset files.

```python
import pandas as pd
```

```python
DATA_DIR = './Data/'
TRACK_FILE = 'spotify_tracks_kaggle_weekly.csv'
ARTIST_FILE = 'spotify_tracks_artist_details.csv'

pd.set_option('display.max_columns', None)
```

### Sample of Tracks Data

```python
tracks_sample = pd.read_csv(DATA_DIR + TRACK_FILE)

print("Sample of tracks data:")
tracks_sample.head()
```

### Sample of Artists Data

```python
artists_sample = pd.read_csv(DATA_DIR + ARTIST_FILE)

print("Sample of artists data:")
artists_sample.head()
```

### Merged Data Sample

```python
merged_sample = pd.merge(tracks_sample, artists_sample, on='track_id', how='left')

print(f"Merged dataset shape: {merged_sample.shape}")
merged_sample.head()
```
